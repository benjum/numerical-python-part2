```python
# Note that you can use "_" in Python numbers for 
# enhanced readability - Python will ignore them

testnum = 1_000_000
```

```python
# Python loop using .iat to set values
def method_loop(n=n):
    df = pd.DataFrame({'a': np.arange(n)})
    for i in range(n):
        df.iat[i, 0] = df.iat[i, 0] + 1
    return df
```

```python
%%time

method_loop(testnum);
```

```python
# List comprehension (build a Python list then assign back)
def method_listcomp(n=n):
    df = pd.DataFrame({'a': np.arange(n)})
    new_col = [x + 1 for x in df['a']]
    df['a_plus1'] = new_col
    return df
```

```python
%%time

method_listcomp(testnum);
```

```python
# Apply a function in a compiled pandas loop internally
def method_map(n=n):
    df = pd.DataFrame({'a': np.arange(n)})
    df['a_plus1'] = df['a'].map(lambda x: x + 1)
    return df
```

```python
%%time

method_map(testnum);
```

```python
# Vectorized pandas/NumPy operation
def method_vectorized(n=n):
    df = pd.DataFrame({'a': np.arange(n)})
    df['a_plus1'] = df['a'] + 1  # vectorized
    return df
```

```python
%%time

method_vectorized(testnum);
```

The vectorized (Pandas/NumPy) approach will be orders of magnitude faster on large datasets
* The operation is executed in compiled C code and uses contiguous memory operations
* Python-level loops (method 1) are the slowest because each iteration involves Python bytecode and DataFrame cell access overhead
* List comprehensions are faster than cell-by-cell loops (they avoid repeated DataFrame cell writes) but still slower than vectorized operations

## What's the data for 'dataset' == 'dino'?  Boolean indexing

It is useful to be able to get elements where certain conditions are true.

Like here, we may want to get only those rows that are part of the 'dino' dataset.

This can be accomplished with boolean indexing, where the index is a True/False condition, and there is one such value for every row.

The following sets up the boolean series of True/False values for every row.

```python
dinodata['dataset'] == 'dino'
```

We can use that as the index to dinodata, i.e., for any dataframe `df` we can use `df[condition]` to get only those rows where `condition` is True

```python
dinodata[dinodata['dataset'] == 'dino']
```

## Plotting

Let's make a scatter plot with only the `dino` dataset

```python
# How do we get that subset of data?

a = dinodata[dinodata['dataset'] == 'dino']
```

Make a plot:

```python
a.plot(x='x', y='y')
```

What's with the zig-zags?

By default, pandas will make a line plot connecting the points, and since the points are plotted out of numerical order, the connecting lines zigs back and forth in the x and y direction.

We actually want to plot this as a scatter plot instead of a line plot.

```python
a.plot(x='x', y='y', kind='scatter')
```

The `kind` parameter makes it very easy to make a variety of different elementary plots:

* `line` : line plot
* `bar` : vertical bar plot
* `barh` : horizontal bar plot
* `hist` : histogram
* `box` : boxplot
* `kde` : kernel density estimation plot
* `density` : same as kde
* `area` : area plot
* `pie` : pie plot
* `scatter` : scatter plot
* `hexbin` : hexbin plot

```python
a.iloc[0:20].plot(x='x', y='y', kind='bar')
```
