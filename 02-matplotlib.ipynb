# pyplot vs figure and axes

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = 3 + np.cos(x)
y2 = 1 + 0.5*np.cos(1+x/0.75)
```

```python
plt.plot(x,y1,x,y2)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y1,x,y2)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x,y1,x,y2)

plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)

ax.plot(x,y1,x,y2)

plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))

ax.plot(x,y1,x,y2)

plt.show()
```

```python
fig,ax = plt.subplots(2,1,figsize=(8,8))

ax[0].plot(x,y1)
ax[1].plot(x,y2)

plt.show()
```

# Creating different types of plots

```python
x = np.linspace(0.5, 3.5, 10)
y = np.cos(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,width=0.2)
plt.show()
```

```python
plt.barh(x,y,height=0.2)
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.linspace(-2, 2, 100)
Xgrid, Ygrid = np.meshgrid(x,y)

f = np.exp(-Ygrid**2) * np.cos(Xgrid)

plt.contourf(f)

plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.linspace(-2, 2, 100)
Xgrid, Ygrid = np.meshgrid(x,y)

f = np.exp(-Ygrid**2) * np.cos(Xgrid)

plt.contourf(Xgrid, Ygrid, f)

plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.linspace(-2, 2, 100)
Xgrid, Ygrid = np.meshgrid(x,y)

f = np.exp(-Ygrid**2) * np.cos(Xgrid)

plt.contourf(Xgrid, Ygrid, f)
ax = plt.gca()
ax.set_xticks([0,2*np.pi,4*np.pi])
ax.set_xticklabels(['0','$2\pi$','$4\pi$'])

plt.show()
```
